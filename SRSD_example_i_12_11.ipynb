# SRSD Benchmark Example (i_12_11)

This notebook demonstrates how SR-KAN was applied to a problem from the SRSD benchmark dataset. The same general pipeline was used for all experiments in the paper, though hyperparameters such as `result_threshold`, `combination_kan_types` and `manipulate_output` were adjusted on a per-problem basis.

```python
import jax.numpy as jnp, jax.random as jr
import jax
from srkan import regressor, SympyEvaluator
import numpy as np
from sklearn.metrics import r2_score
jax.config.update("jax_enable_x64", True)
```

```python
x = jnp.array(np.loadtxt("YOUR_TRAIN_DATA.txt"))
y = x[:,[-1]]
x = x[:,:-1]
```

```python
ex = regressor(jr.key(14), 
                result_threshold=1e-10,
                activation_function=jax.nn.gelu,
                manipulate_output=[],combination_kan_types=[["mult","mult","mult"]],
                functions=["all"],verbosity=1).fit(x,y)
print(str(ex))
print(SympyEvaluator(ex)(x,y))
print(r2_score(y, SympyEvaluator(ex)(x, None, mse=False)))
```
